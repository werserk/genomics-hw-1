### Решение (База)

> Не делал в Latex, тк это не изолированная задача.

> **Цепь Маркова.** Обозначим через
>
>
> $$
> X_t = (S^A_t, S^B_t)
> $$
>
> счёт после $t$-го розыгрыша, где $S^A_t$ — очки Алексея, $S^B_t$ — очки Бориса. Тогда $\{X_t\}_{t\ge 0}$ — однородная цепь Маркова на пространстве достижимых счётов, старт $X_0=(0,0)$.
>
> **Пространство состояний $\mathcal S$** — все пары $(i,j)$ очков, достижимые по правилам:
> - до «деюса» это $0\le i,j\le 21$ и игра останавливается при $(21,j)$ или $(i,21)$ при втором счёте $\le 19$;
> - после попадания в $(20,20)$ возможны состояния $(20+a,20+b)$ для целых $a,b\ge 0$ до момента, когда $|a-b|=2$ (то есть разница в счёте 2 очка).
>
> **Терминальные состояния**:
> - победа Алексея: $\{(21,j): 0\le j\le 19\}\ \cup\ \{(20+k,20+k-2): k\ge 2\}$;
> - победа Бориса: $\{(i,21): 0\le i\le 19\}\ \cup\ \{(20+k-2,20+k): k\ge 2\}$.
>
> Из любого такого состояния игра уже не продолжается.
>
> **Переходные вероятности.** Если $(i,j)$ — *нетерминальное* состояние, то следующий розыгрыш добавляет 1 очко победителю розыгрыша:
>
> $$
> \mathbb P\bigl(X_{t+1}=(i+1,j)\mid X_t=(i,j)\bigr)=p,\qquad
> \mathbb P\bigl(X_{t+1}=(i,j+1)\mid X_t=(i,j)\bigr)=1-p.
> $$
>
> Если $(i,j)$ терминальное, то это поглощающее состояние:
>
> $$
> \mathbb P\bigl(X_{t+1}=(i,j)\mid X_t=(i,j)\bigr)=1.
> $$
>
> **Инвариантные распределения.** В цепи есть множество замкнутых классов сообщающихся состояний: каждый терминальный счёт $s\in\mathcal A$ (поглощающее состояние) образует свой класс $\{s\}$. Поэтому **любое** распределение, целиком поддержанное на множестве поглощающих состояний $\mathcal A$, является инвариантным:
>
> $$
> \pi P = \pi\quad \Longleftrightarrow\quad \pi(\mathcal A)=1.
> $$
>
> В частности, $\delta_s$ (точечная масса в конкретном финальном счёте) — инвариантна для каждого поглощающего состояния $s$.
>
> **Эргодичность.** Цепь **не эргодическая**:
> - она **не является неразложимой (irreducible)**, т.к. имеет много замкнутых классов (каждый финальный счёт — отдельный класс);
> - следовательно, **не существует единственного** стационарного распределения, и предельное поведение зависит от начального состояния.
>
> При этом из $(0,0)$ цепь почти наверняка за конечное время попадает в $\mathcal A$ (игра заканчивается), т.е. это поглощающая цепь с транзиентными нетерминальными состояниями.

Постройте алгоритм оценки параметра $p$ с помощью метода максимального правдоподобия по данным. Выведите оценку ниже:

### Решение (Оценка $p$)

>Пусть дан набор игр (траекторий) $X^{(g)}$, где каждая игра — последовательность исходов розыгрышей длины $n_g$:
>
>$$
>X^{(g)}_1,\dots,X^{(g)}_{n_g},\qquad X^{(g)}_t\in\{A,\,B\}
>$$
>
>Считаем, что розыгрыши независимы и $\mathbb P(X=A)=p$, $\mathbb P(X=B)=1-p$
>
>Обозначим общее число выигранных розыгрышей Алексеем и Борисом по всему датасету:
>
>$$
>N_A = \sum_g \sum_{t=1}^{n_g} \mathbf 1\{X^{(g)}_t=A\},\qquad
>N_B = \sum_g \sum_{t=1}^{n_g} \mathbf 1\{X^{(g)}_t=B\}.
>$$
>
>Тогда правдоподобие (с точностью до порядка в последовательностях, который тут неважен) равно
>
>$$
>L(p)=p^{N_A}(1-p)^{N_B},
>$$
>
>a лог-правдоподобие $\ell(p)=N_A\log p+N_B\log(1-p)$
>
>Максимум на $(0,1)$ находится из условия $\ell'(p)=0$:
>
>$$
>\frac{N_A}{p}-\frac{N_B}{1-p}=0\ \Longrightarrow\ \hat p=\frac{N_A}{N_A+N_B}.
>$$
>
>Это просто **доля розыгрышей, выигранных Алексеем**, по всему датасету
>
>- Если данные закодированы как 0/1 (например, 1 = Алексей, 0 = Борис), то $\hat p$ — это среднее по всем наблюдениям: $\hat p=\frac>{1}{N}\sum X$.
>- Граничные случаи: если $N_A=0$, то $\hat p=0$; если $N_B=0$, то $\hat p=1$
>
>**Алгоритм ММП:** пройти по всем играм, посчитать суммарное число розыгрышей и суммарное число побед Алексея, вернуть их отношение
>

>
>$$
>N=37016,\quad N_A=20466,\quad \hat p=0.5528960449535336
>$$

### Эксперименты (15 баллов)

Теперь займёмся (эээ)экспериментами. Для исследования данной модели нам понадобится отдельный класс.

```python
class BadmintonGenerator:
    def __init__(self, p: float, players: list[str] = ["Boris", "Alexey"]):
        if not (0.0 <= p <= 1.0):
            raise ValueError(f"p must be in [0, 1], got {p}")
        self.p = p  # win proba of Alexey
        self.players = players

    def simulate(self, NTrajs: int = 10) -> dict[str, object]:

        games: list[list[int]] = []  # each game is a list of outcomes 0-1

        for _ in tqdm.tqdm(range(NTrajs)):
            game: list[int] = []
            score_a = 0  # Alexey
            score_b = 0  # Boris

            while True:
                alexey_wins = np.random.random() < self.p
                game.append(1 if alexey_wins else 0)

                if alexey_wins:
                    score_a += 1
                else:
                    score_b += 1

                if (score_a >= 21 or score_b >= 21) and abs(score_a - score_b) >= 2:
                    break

            games.append(game)

        return {"players": copy(self.players), "games": games}

    def fit(self, games: list[list[int]]) -> float:

        n_total = sum(len(g) for g in games)
        if n_total == 0:
            raise ValueError("Cannot fit p: empty dataset (0 rallies).")

        n_alexey = sum(sum(g) for g in games)
        self.p = n_alexey / n_total
        return self.p
```

Реализуйте ниже функции для получения последовательности счетов (нам понадобится ниже), а также вычисления исхода игры.

```python
def getScores(game):

    scores = [[0, 0]]
    for outcome in game:
        if outcome not in (0, 1):
            raise ValueError(f"Bad outcome value: {outcome} (expected 0 or 1)")
        s0, s1 = scores[-1]
        if outcome == 0:
            scores.append([s0 + 1, s1])
        else:
            scores.append([s0, s1 + 1])

    return scores


def getWinner(game, players):

    score1 = int(sum(game))
    score0 = int(len(game) - score1)

    if score0 == score1:
        # Should not happen for a valid finished badminton game, but keep it safe.
        return None, (score0, score1)

    winner_idx = 0 if score0 > score1 else 1
    return players[winner_idx], (score0, score1)
```

## Исследуем, как играют игроки

Оценённый нами параметр $p$ ничего не говорит о более сложных вероятностях. Их можно пытаться вычислять аналитически, но гораздо интереснее придумать, как это можно примерно посчитать с помощью экспериментов, потому что не всегда модель достаточно простая, чтобы выписывать аналитику.

Нас будет интересовать несколько объектов, связанных с моделью:

1. Вероятность победы каждого из игроков.
2. Распределение длительности игр.
3. Распределение финальных счётов (нас будет интересовать приближённый закон).
4. Вероятность возникновения ситуации 20-20.

Для начала реализуем все эти вычисления в виде одного класса эксперимента.

```python
class BadmintonExperiment:

    def __init__(self, p):
        self.gen = BadmintonGenerator(p)

    def getScores(self, game):
        return getScores(game)

    def getWinner(self, game):
        return getWinner(game, self.gen.players)

    def performExperiment(self, NTrials, returnExactGames=False):
        gameData = self.gen.simulate(NTrajs=NTrials)
        games = gameData["games"]
        wins0 = 0
        wins1 = 0

        lengths = [len(g) for g in games]
        max_len = max(lengths) if lengths else 0
        gLCounts = np.zeros(max_len + 1, dtype=np.int64)

        # final score distribution (Boris on Y axis, Alexey on X axis)
        final_scores = []

        cnt_2020 = 0

        for g in games:
            L = len(g)
            gLCounts[L] += 1

            score1 = int(sum(g))
            score0 = int(L - score1)
            final_scores.append((score0, score1))

            if score0 > score1:
                wins0 += 1
            else:
                wins1 += 1

            s0 = 0
            s1 = 0
            hit = False
            for outcome in g:
                if outcome == 0:
                    s0 += 1
                else:
                    s1 += 1
                if s0 == 20 and s1 == 20:
                    hit = True
                    break
            if hit:
                cnt_2020 += 1

        gLProbas = gLCounts / float(len(games))
        winningProbas = np.array([wins0, wins1], dtype=float) / float(len(games))
        proba2020 = cnt_2020 / float(len(games))

        if final_scores:
            max_score = max(max(s0, s1) for s0, s1 in final_scores)
        else:
            max_score = 0
        finalScoresDist = np.zeros((max_score + 1, max_score + 1), dtype=float)
        for s0, s1 in final_scores:
            finalScoresDist[s0, s1] += 1.0
        if len(games) > 0:
            finalScoresDist /= float(len(games))

        if returnExactGames:
            return {
                "gameLengthDistribution": gLProbas,
                "finalScoreDistribution": finalScoresDist,
                "winningProbas": winningProbas,
                "proba2020": proba2020,
            }, gameData
        return {
            "gameLengthDistribution": gLProbas,
            "finalScoreDistribution": finalScoresDist,
            "winningProbas": winningProbas,
            "proba2020": proba2020,
        }
```

Проверим, что всё работает на простом примере.

```python
bex = BadmintonExperiment(p=0.2)

expLog = bex.performExperiment(NTrials=10000, returnExactGames=False)
```

```python
def plotExperiment(expLog):

    f, (ax1,ax2) = plt.subplots(1,2, figsize=(14,5))

    ax1.grid()
    ax1.set_title("Распределение длительностей игр", fontsize=16)
    ax1.set_xlabel("Длительность", fontsize=16)
    ax1.set_ylabel("Вероятность", fontsize=16)
    ax1.bar(np.arange(len(expLog["gameLengthDistribution"])), expLog["gameLengthDistribution"])

    ax2.grid()
    ax2.set_title("Лог-вероятности финального счёта", fontsize=16)
    ax2.set_xlabel(f"Игрок {bex.gen.players[1]}", fontsize=16 )
    ax2.set_ylabel(f"Игрок {bex.gen.players[0]}", fontsize=16)
    im=ax2.imshow(np.log( expLog["finalScoreDistribution"] +1e-5))

    cbar_ax = f.add_axes([0.95, 0.15, 0.05, 0.7])
    f.colorbar(im, cax=cbar_ax)
```

```python
plotExperiment(expLog)
```

### Сопоставим матожидание длины игры с теорией (10 баллов)

Матожидание длины игры можно посчитать аналитически, как это сделать, хорошо показывает математическая задача из начала ноутбука, но тут для бумаги формулы не такие приятные. Мы попробуем составить систему линейных уравнений и решить её с помощью ``np.linalg``.

Удобнее всего поделить игру на две части: игра до момента $20:20$ и после момента $20:20$.

Найдите матожидание длины игры при условии, что игра начинается в счёте $(0,0)$ и что игра заканчивается в случаях, когда один из игроков достигает $21$ или на табло становится $20:20$

> Обозначим через $T_{i,j}$ матожидание числа оставшихся розыгрышей **до момента остановки** в укороченной игре, которая заканчивается либо при достижении кем-то счёта $21$ (до деюса), либо при попадании в $20{:}20$.
> 
> Тогда для всех нетерминальных состояний $(i,j)$ выполняется уравнение первого шага:
> 
> $$
> T_{i,j}=1 + p\,T_{i+1,j} + (1-p)\,T_{i,j+1}.
> $$
> 
> Граничные условия (остановка):
> 
> - $T_{20,20}=0$;
> - $T_{21,j}=0$ для $0\le j\le 19$;
> - $T_{i,21}=0$ для $0\le i\le 19$.
> 
> Чтобы собрать полный ответ для исходной игры, также нужна вероятность дойти до $20{:}20$ до остановки. Обозначим её $q_{i,j}=\mathbb P(\text{дойдём до }20{:}20\mid (i,j))$. Тогда:
> 
> $$
> q_{i,j}= p\,q_{i+1,j} + (1-p)\,q_{i,j+1},
> $$
> 
> с граничными условиями $q_{20,20}=1$, $q_{21,j}=0$ (для $j\le 19$), $q_{i,21}=0$ (для $i\le 19$).

```python
def _build_state_index():
    states = []
    for i in range(21):
        for j in range(21):
            if i == 20 and j == 20:
                continue
            states.append((i, j))
    idx = {s: k for k, s in enumerate(states)}
    return states, idx


def expected_until_21_or_2020(p: float):
    states, idx = _build_state_index()
    n = len(states)

    A = np.zeros((n, n), dtype=float)
    b = np.zeros(n, dtype=float)

    for s_id, (i, j) in enumerate(states):
        if (i == 21 and j <= 19) or (j == 21 and i <= 19) or (i == 20 and j == 20):
            A[s_id, s_id] = 1.0
            b[s_id] = 0.0
            continue

        A[s_id, s_id] = 1.0
        b[s_id] = 1.0

        ni, nj = i + 1, j
        mi, mj = i, j + 1

        if ni == 20 and nj == 20:
            pass
        elif ni == 21 and nj <= 19:
            pass
        elif ni <= 20 and nj <= 20:
            A[s_id, idx[(ni, nj)]] -= p

        if mi == 20 and mj == 20:
            pass
        elif mj == 21 and mi <= 19:
            pass
        elif mi <= 20 and mj <= 20:
            A[s_id, idx[(mi, mj)]] -= (1 - p)

    T = np.linalg.solve(A, b)

    Aq = np.zeros((n, n), dtype=float)
    bq = np.zeros(n, dtype=float)

    for s_id, (i, j) in enumerate(states):
        Aq[s_id, s_id] = 1.0

        ni, nj = i + 1, j
        mi, mj = i, j + 1

        if ni == 20 and nj == 20:
            bq[s_id] += p * 1.0
        elif ni == 21 and nj <= 19:
            bq[s_id] += p * 0.0
        else:
            Aq[s_id, idx[(ni, nj)]] -= p

        if mi == 20 and mj == 20:
            bq[s_id] += (1 - p) * 1.0
        elif mj == 21 and mi <= 19:
            bq[s_id] += (1 - p) * 0.0
        else:
            Aq[s_id, idx[(mi, mj)]] -= (1 - p)

    q = np.linalg.solve(Aq, bq)

    return float(T[idx[(0, 0)]]), float(q[idx[(0, 0)]])


T00, q00 = expected_until_21_or_2020(gen.p)
T00, q00
```

Теперь найдите ожидаемую продолжительность игры от момента $20:20$. Здесь удобнее всего построить искусственную цепь Маркова с состояниями $0:0,0:1,1:0,2:0,0:2$. 

> После $20{:}20$ удобно рассматривать разность $D_t = S^A_t - S^B_t$ относительно счёта $20{:}20$.
> 
> В момент $20{:}20$ имеем $D_0=0$. Каждый следующий розыгрыш увеличивает $D$ на $+1$ с вероятностью $p$ и уменьшает на $-1$ с вероятностью $1-p$. Игра после деюса заканчивается при $D=+2$ (Алексей выиграл с разницей 2) или $D=-2$ (Борис выиграл с разницей 2).
> 
> Обозначим $E_d$ — матожидание числа оставшихся розыгрышей до поглощения, если текущая разность равна $d\in\{-2,-1,0,1,2\}$. Тогда:
> 
> - граничные условия: $E_{-2}=E_{2}=0$;
> - уравнения первого шага:
> 
> $$
> E_0 = 1 + p E_{1} + (1-p) E_{-1},
> $$
> 
> $$
> E_1 = 1 + p E_{2} + (1-p) E_{0},
> \qquad
> E_{-1} = 1 + p E_{0} + (1-p) E_{-2}.
> $$
> 
> Решив эту систему, получаем $E_0$ — ожидаемую длину игры после $20{:}20$.

```python
def expected_after_2020(p: float) -> float:
    return 2.0 / (p * p + (1 - p) * (1 - p))

E0 = expected_after_2020(gen.p)
E0
```

> Собираем полный ответ:
> 
> - $T_{0,0}$ — ожидаемое число розыгрышей до остановки в укороченной игре (до $21$ или до $20{:}20$);
> - $q_{0,0}$ — вероятность дойти до $20{:}20$;
> - $E_0$ — ожидаемое число розыгрышей после $20{:}20$ до конца игры.
> 
> Тогда матожидание полной длины игры из $(0,0)$:
> 
> $$
> \mathbb E[\text{длина}] = T_{0,0} + q_{0,0}\cdot E_0.
> $$

```python
E_theory = T00 + q00 * E0

try:
    exp_for_p = expLogs.get(1.0, None)
except NameError:
    exp_for_p = None

if exp_for_p is None:
    exp_for_p = BadmintonExperiment(p=gen.p).performExperiment(NTrials=100000, returnExactGames=False)

gL = exp_for_p["gameLengthDistribution"]
E_exp = float(np.sum(gL * np.arange(len(gL))))

E_theory, E_exp, abs(E_theory - E_exp)
```

> В нашем сравнении получилось $\mathbb E_{\text{theory}}\approx 36.3948$ и $\mathbb E_{\text{exp}}\approx 36.38577$ (разница $\approx 0.0090$), то есть совпадение на уровне $\sim 10^{-2}$ розыгрыша
