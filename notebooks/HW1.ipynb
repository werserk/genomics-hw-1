Для набора $p=0.1, 0.2, 0.25,0.3,0.35,0.4,0.425,0.45,0.475, 0.48, 0.49, 0.495, 0.5$ посчитайте экспериментально среднее и дисперсию длительности одной партии и нарисуйте графики среднего и дисперсии в зависимости от выбранного параметра $p$.

При малых $p$ оценки делаются сложнее, поэтому во всех экспериментах используйте минимум ``NTrials=1e5``.

```python
ps = [0.1,0.2,0.3,0.35,0.4,0.425,0.45,0.475,0.48,0.49,0.495,0.5]
means = np.zeros(len(ps))
variances = np.zeros(len(ps))
for pId in range(len(ps)):
    bexx = BadmintonExperiment(ps[pId])
    expData = bexx.performExperiment(NTrials=100000)
    means[pId]=np.sum(expData["gameLengthDistribution"]*np.arange(len(expData["gameLengthDistribution"])))
    variances[pId]=np.sum(expData["gameLengthDistribution"]*(np.arange(len(expData["gameLengthDistribution"]))**2)) - means[pId]**2
```

```python
f, (ax1,ax2) = plt.subplots(1,2,figsize=(14,7))

ax1.grid()
ax1.set_title("Средняя длительность", fontsize=16)
ax1.set_xlabel("p", fontsize=16)
ax1.plot(np.array(ps), means)

ax2.grid()
ax2.set_title("Дисперсия длительности", fontsize=16)
ax2.set_xlabel("p", fontsize=16)
ax2.plot(np.array(ps), variances)
```
